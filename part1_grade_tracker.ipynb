```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

cleaned_students = []

for raw in raw_students:
    # basic cleaning
    cleaned_name = raw["name"].strip().title()
    cleaned_roll = int(raw["roll"])

    # marks: split on "," and strip spaces, then to int
    parts = raw["marks_str"].split(",")
    cleaned_marks = [int(p.strip()) for p in parts]

    # name validation
    words = cleaned_name.split()
    only_alpha = True
    for w in words:
        if not w.isalpha():
            only_alpha = False
            break

    status = "✓ Valid name" if only_alpha else "✗ Invalid name"
    print(status)

    # profile card
    print("================================")
    print(f"Student : {cleaned_name}")
    print(f"Roll No : {cleaned_roll}")
    print(f"Marks   : {cleaned_marks}")
    print("================================")

    cleaned_students.append({
        "name": cleaned_name,
        "roll": cleaned_roll,
        "marks": cleaned_marks
    })

# lookup using for/else (different style)
for student in cleaned_students:
    if student["roll"] == 103:
        n = student["name"]
        print(n.upper())
        print(n.lower())
        break
else:
    print("Student with roll 103 not found.")
```

```python
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

def get_grade(mark):
    if 90 <= mark <= 100:
        return "A+"
    elif 80 <= mark <= 89:
        return "A"
    elif 70 <= mark <= 79:
        return "B"
    elif 60 <= mark <= 69:
        return "C"
    else:
        return "F"

print(f"Marks report for {student_name}")
print("--------------------------------")

# Part 1: Grades
for subject, mark in zip(subjects, marks):
    grade = get_grade(mark)
    print(f"{subject}: {mark} → {grade}")

# Part 2: Analysis
total_marks = sum(marks)
average_marks = total_marks / len(marks)

highest_mark = max(marks)
lowest_mark  = min(marks)

highest_subject = subjects[marks.index(highest_mark)]
lowest_subject  = subjects[marks.index(lowest_mark)]

print("\nSummary")
print("-------")
print(f"Total marks   : {total_marks}")
print(f"Average marks : {round(average_marks, 2)}")
print(f"Highest       : {highest_subject} {highest_mark}")
print(f"Lowest        : {lowest_subject} {lowest_mark}")

# Part 3: While loop input
print("\n=== Add New Subjects (type 'done' to finish) ===")
new_subjects_count = 0

while True:
    subject = input("Enter subject name (or 'done' to stop): ").strip()

    if subject.lower() == "done":
        break

    marks_input = input(f"Enter marks for {subject} (0-100): ").strip()

    if not marks_input.isdigit():
        print("Invalid input! Please enter numbers only.")
        continue

    mark = int(marks_input)

    if mark < 0 or mark > 100:
        print("Marks should be between 0 and 100.")
        continue

    subjects.append(subject)
    marks.append(mark)
    new_subjects_count += 1
    print(f"Added {subject} with {mark} marks.")

print("\nNew subjects added:", new_subjects_count)

updated_average = sum(marks) / len(marks)
print("Updated average marks:", round(updated_average, 2))
```

```python
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

def calculate_average(marks_list):
    return round(sum(marks_list) / len(marks_list), 2)

print("Name              | Average | Status")
print("----------------------------------------")

pass_count = 0
fail_count = 0
total_avg_sum = 0.0
topper_name = ""
topper_avg = 0.0

for name, marks in class_data:
    avg = calculate_average(marks)

    if avg >= 60:
        status = "Pass"
        pass_count += 1
    else:
        status = "Fail"
        fail_count += 1

    print(f"{name:<18} | {avg:^7} | {status}")

    if avg > topper_avg:
        topper_avg = avg
        topper_name = name

    total_avg_sum += avg

print("\nSummary")
print("-------")
print("Passed students :", pass_count)
print("Failed students :", fail_count)
print(f"Topper          : {topper_name} ({topper_avg})")

class_avg = total_avg_sum / len(class_data)
print("Class Average   :", round(class_avg, 2))
```

```python
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

def print_step(label, value):
    print(label)
    print(value)

# Step 1: strip
clean_essay = essay.strip()
print_step("1. Clean Essay:", clean_essay)

# Step 2: title case
title_case = clean_essay.title()
print()
print_step("2. Title Case:", title_case)

# Step 3: count of 'python'
python_count = clean_essay.count("python")
print()
print_step("3. Count of 'python':", python_count)

# Step 4: replace 'python' with 'Python 🐍'
replaced_text = clean_essay.replace("python", "Python 🐍")
print()
print_step("4. Replaced Text:", replaced_text)

# Step 5: split into sentences
sentences = clean_essay.split(". ")
print()
print_step("5. Sentences List:", sentences)

# Step 6: numbered sentences
print()
print("6. Numbered Sentences:")
for idx, sentence in enumerate(sentences, start=1):
    if not sentence.endswith("."):
        sentence = sentence + "."
    print(f"{idx}. {sentence}")
```
